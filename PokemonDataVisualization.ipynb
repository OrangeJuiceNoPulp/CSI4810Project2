```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pokemon_text_cleaned.csv")
```

```python
len(df)
```

```python
classes = list(df.columns)
classes.remove('DexNumber')
classes.remove('Name')
classes.remove('FlavorText')
```

```python
outputs = []
for text in df['FlavorText']:
    outputs.append(len(text))

plt.hist(outputs, bins=10, color='c')

plt.title('Pokedex Entry Character Length (All Generations)')
plt.ylabel('Number of Pokedex Entries')
plt.xlabel('Length of Pokedex Entry')
plt.xticks(range(0,251,25))
plt.show()
```

```python
for i in classes:
    plt.bar(i, df[i].sum(), color='c')

plt.title('Pokedex Entry Count By Pokemon Type (All Generations)')
plt.ylabel('Number of Pokedex Entries')
plt.xlabel('Type')
plt.xticks(rotation=90)
plt.show()
```

```python
for i in classes:
    plt.bar(i, df[df['DexNumber'] < 152][i].sum(), color='c')

plt.title('Pokedex Entry Count By Pokemon Type (Generation 1)')
plt.ylabel('Number of Pokedex Entries')
plt.xlabel('Type')
plt.xticks(rotation=90)
plt.show()
```
