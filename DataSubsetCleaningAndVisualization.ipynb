```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pokemon_text_cleaned.csv")

subset = pd.DataFrame({'DexNumber':df['DexNumber'], 'Name':df['Name'], 'FlavorText':df['FlavorText'], 'Grass':df['Grass'], 'Water':df['Water'], 'Fire':df['Fire']})
```

```python
classes = ['Grass', 'Water', 'Fire']
test_array = np.zeros_like(np.asarray(subset['Grass']))
for i,name in enumerate(classes):
    test_array = test_array + np.asarray(subset[name])
subset = subset[pd.Series(test_array) == 1]
```

```python
subset.to_csv('pokemon_3_types_subset.csv', index=False)
```

```python
df = pd.read_csv("pokemon_3_types_subset.csv")
```

```python
len(df)
```

```python
outputs = []
for text in df['FlavorText']:
    outputs.append(len(text))

plt.hist(outputs, bins=10, color='c')

plt.title('Pokedex Entry Character Length (Grass, Water, Fire)')
plt.ylabel('Number of Pokedex Entries')
plt.xlabel('Length of Pokedex Entry')
plt.xticks(range(0,251,25))
plt.show()
```

```python
for i in classes:
    plt.bar(i, df[i].sum(), color='c')

plt.title('Pokedex Entry Count By Pokemon Type (Grass, Water, Fire)')
plt.ylabel('Number of Pokedex Entries')
plt.xlabel('Type')
plt.xticks(rotation=90)
plt.show()
```
